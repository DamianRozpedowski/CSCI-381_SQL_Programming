**Query 1 - Return orders placed in June 2015**

**Proposition:** Retrieve the order details for all orders placed in June 2015. 

**Tables:** <span style="color: var(--vscode-foreground);">Sales.[Order]</span>

**Columns:** orderid, orderdate, customerid, and employeeid.

**Predicate:** Orders must have been placed within the month of June 2015, meaning and date between June 1st, 2015 and the end of June (June 30th, 2015). This can be accomplished using a simple WHERE clause.

```sql
USE TSQLV4
SELECT orderid, orderdate, custid, empid
FROM Sales.Orders
WHERE orderdate >= '2015-06-01' AND orderdate < '2015-07-01';


USE Northwinds2022TSQLV7
SELECT orderid, orderdate, customerid, employeeid
FROM Sales.[Order]
WHERE orderdate >= '20150601' AND orderdate < '20150701';

SELECT orderid, orderdate, customerid, employeeid
FROM Sales.[Order]
WHERE orderdate BETWEEN '2015-06-01' AND '2015-06-30';
```

**Query 2 - Return orders placed on the last day of the month**

**Proposition:** Retrieve the order details for all orders placed on the last day of the month.

<span style="color: var(--vscode-foreground);"><b>Tables:&nbsp;</b></span>    <span style="color: var(--vscode-foreground);">Sales.[Order]</span>

<span style="color: var(--vscode-foreground);"><b>Columns:</b></span> orderid, orderdate, customerid, and employeeid.

**Predicate:** Orders must have been placed on the last day of the month. We can accomplish this by using the WHERE clause. Since orderdate only stores the day date (YYYY-MM-DD) we can simply use EOMONTH function on our orderdate and compare that with our orderdate to see whether it was placed at the end of the month.

```sql
USE TSQLV4
SELECT orderid, orderdate, custid, empid
FROM Sales.Orders
WHERE orderdate = EOMONTH(orderdate);


USE Northwinds2022TSQLV7
SELECT orderid, orderdate, customerid, employeeid
FROM Sales.[Order]
WHERE orderdate = EOMONTH(orderdate);
```

**Query 4 - Return order details with total value (qty\*unitprice) greater than 10000 sort by total value.**

**Proposition:** Retrieve the orderid and totalvalue for <span style="color: var(--vscode-foreground);">orders with a total value (quantity*unitprice) greater than 10000, sort by total value.</span>

**<span style="color: var(--vscode-foreground);">Tables:&nbsp;</span>**   <span style="color: var(--vscode-foreground);">Sales.OrderDetail</span>

**<span style="color: var(--vscode-foreground);">Columns:</span>** <span style="color: var(--vscode-foreground);">OrderId, ProductId, UnitPrice, Quantity. For this proposition we</span> want a desired output which includes OrderId and TotalValue which we must obtain by doing SUM(quantity\*unitprice) as TotalValue so that the same orderid's are being combined.

**Predicate:** Orders must have exceeded a total value of 10000.

```sql
USE TSQLV4
SELECT orderid, SUM(qty * unitprice) AS totalvalue
FROM Sales.OrderDetails
GROUP BY orderid
HAVING SUM(qty * unitprice) > 10000
ORDER BY totalvalue DESC;


USE Northwinds2022TSQLV7
SELECT orderid, SUM(quantity * unitprice) AS totalvalue
FROM Sales.OrderDetail
GROUP BY orderid
HAVING SUM(quantity * unitprice) > 10000
ORDER BY totalvalue DESC;
```
